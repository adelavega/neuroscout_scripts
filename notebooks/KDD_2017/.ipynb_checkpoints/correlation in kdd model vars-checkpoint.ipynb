```python
import pandas as pd
import seaborn as sns
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from coda import EventTransformer, BIDSEventReader
from glob import glob
```

```python
from modulefindere
```

```python
reader = BIDSEventReader()
all_runs = pd.concat([reader.read(f) for f in glob('scratch/*.tsv')])

transformer = EventTransformer(all_runs)
data = transformer.data#.replace([np.inf, -np.inf], np.nan)

data = data.drop('onset', axis=1)
#data = data.fillna('0')
data = data.convert_objects(convert_numeric=True)
```

```python
corr = data.corr()
```

```python
corr = corr.round(2)
```

```python
data.long_freq.mean()
```

```python
# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
```

```python
width = 8
height = 7
plt.figure(figsize=(width, height))
sns.heatmap(corr, mask=mask, annot=True)
```
