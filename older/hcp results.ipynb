```python
import nibabel as nib
from nilearn import plotting as nip
import glob
%matplotlib inline
from matplotlib.colors import LinearSegmentedColormap
import seaborn as sns
```

```python
blues = LinearSegmentedColormap.from_list('blue', sns.color_palette("Blues"))
```

## Clarifai labels

```python
zstats = glob.glob('hcp/clarifai_fx/group/stats/c*/*1_thre*')
```

```python
conditions = ['street', 'outdoors', 'light', 'adult']
contrasts = [['street', 'T', conditions, [1, 0, 0, 0]],
             ['outdoors', 'T', conditions, [0, 1, 0, 0]],
             ['light', 'T', conditions, [0, 0, 1, 0]],
             ['adult', 'T', conditions, [0, 0, 0, 1]],
             ['street > outdoors', 'T', conditions, [1, -1, 0, 0]],
             ['street > light', 'T', conditions, [1, 0, -1, 0]],
             ['street > adult', 'T', conditions, [1, 0, 0, -1]],
             ['outdoors > light', 'T', conditions, [0, 1, -1, 0]],
             ['outdoors > adult', 'T', conditions, [0, 1, 0, -1]],
             ['light > adult', 'T', conditions, [0, 0, 1, -1]]
             ]
```

```python
names = list(zip(*contrasts))[0]
for i, f in enumerate(zstats):
    nip.plot_stat_map(nib.load(f), title=names[i])
```
